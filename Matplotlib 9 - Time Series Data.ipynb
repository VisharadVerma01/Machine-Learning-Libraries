```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.dates as mpl_dates
```

### Time Series Data

```python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y, linestyle='dotted')
#Can use marker=none if you do not want markers
#Plot_date is how to plot time series data
#Plot_date default is a scatter plot, need to define a 
##linestyle to actually plot it

plt.gcf().autofmt_xdate()
#Fits the date properly below the x axis

date_format = mpl_dates.DateFormatter('%b, %d, %Y')
plt.gca().xaxis.set_major_formatter(date_format)
#Default date is yyyy-mm-dd, using the above line
##will make it MMMM-dd-yyyy

plt.tight_layout()

plt.show()
```

### Real World Data 

```python
plt.style.use('ggplot')

data = pd.read_csv('ts.csv')

data['Date'] = pd.to_datetime(data['Date'])
#Converts data from string to date
data.sort_values('Date',inplace=True)
#sorts the date ; inplace autoassigns it

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date,price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
