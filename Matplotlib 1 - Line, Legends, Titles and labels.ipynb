```python
import matplotlib.pyplot as plt
```

#### Making a line

```python
x = [1,2,3] #Defining the x-coordinate
y = [5,7,4] #Defining the y-coordinate

x2 = [1,2,3] #Defining the x2-coordinate
y2 = [10,14,12] #Defining the y2-coordinate

plt.plot(x2,y2, label = 'First Line') #Need a label for legend
plt.plot(x,y, label = 'Second Line') #Plotting the graph
plt.xlabel("Plot Number") #X-Label
plt.ylabel("Important var") #Y-Label

plt.title('Interesting Graph\nCheck it out') #Title of the graph

plt.legend() #Display a legend i.e. the small information box on the right 
             #which displays the labels.

plt.show() #Showing the graph
```
