```python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

### Real-Time Data Visualization

```python
%matplotlib notebook
#For updating the graph
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()
#Calling the count function

def animate(i): 
    x_vals.append(next(index)) #Counting the times animate is run, x will look like [1,2,3..]
    y_vals.append(random.randint(0, 5)) #Appends a random no between 0 and 5 to y
    
    plt.cla() #Clears the axis
    plt.plot(x_vals, y_vals) #Plots the values
    
ani = FuncAnimation(plt.gcf(), animate, interval=1000)
#Animation function for smootness and updating
#Calls the animate function to draw the plot
#Gets the figure using plt.gfc()
#Updates every 1 second

plt.tight_layout()
plt.show()
```

### Real time example

```python
%matplotlib notebook
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()

def animate(i):
    data = pd.read_csv('data.csv') #Takes values from an infinitely generating csv file
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']
    
    plt.cla()
    plt.plot(x, y1, label="Channel 1")
    plt.plot(x, y2, label="Channel 2")
    
    plt.legend(loc='upper left')
    plt.tight_layout()
    
ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()

#Cannot see this in Jupyter as Jupyter uploads an instance of the data
```
