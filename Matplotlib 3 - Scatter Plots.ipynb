```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Scatter Plots

```python
x = [1,2,3,4,5,6,7,8]
y = [2,3,5,2,7,5,4,2]

plt.scatter(x,y,label='Scatter',color="#78128a", marker="4", s=50)
#marker = Types of Markers
#s = Size of the markers
#plt.scatter makes a scatter plot

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph')
plt.legend()

plt.show()
```

### Another example of scatter plots

```python
plt.style.use('ggplot')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
#Defines a colormap, for example, (5,7) will be marked as 7 i.e. high
##or dark shade of green

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
#Defines a sizemap, for example, (7,4) will have a size of 486

plt.scatter(x,y,s=sizes, c=colors,cmap='Greens', edgecolor='black',
            linewidth=1, alpha=0.75)

#cmap - Defines a color for the colormap, we chose the color green here
#Edgecolor - Makes the points more defined
#linewidth - Makes the border more T H I C C

cbar = plt.colorbar()
#Call the method and assign it to a variable
cbar.set_label('Satisfaction')
#Sets a label to the colorbar

plt.tight_layout()

plt.show()
```

### Real World Example of Scatter Plots

```python
plt.style.use('ggplot')

data = pd.read_csv('sp.csv')

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count,likes, c=ratio, cmap='summer',
            edgecolor='black',linewidth=1, alpha=0.75)
#summer is a color scheme

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')
#Since we found an outlier, all the data was cramped
#Using log, the data is more distributed

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
