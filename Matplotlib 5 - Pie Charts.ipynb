```python
import matplotlib.pyplot as plt
```

### Pie Charts

```python
days = [1,2,3,4,5]

sleeping = [7,8,11,6,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

slices = [7,2,2,13]
#Slices denote the ratio/share of the circle they take up
activities = ['sleeping','eating','working','playing']
cols = ['#4a75ba','#a834bf','#b81c21','#0e1661']

plt.pie(slices,labels=activities, colors=cols, startangle=90, shadow=True,
        explode = (0,0.1,0,0), autopct='%1.1f%%')

#startangle - What angle the pie chart starts from
#shadow - Adds a 3D effect
#Explode - Takes a share out for visual attention
#autopct - Shows the percentage taken up

#plt.xlabel('x')
#plt.ylabel('y')
plt.title('Interesting Graph')
#plt.legend()

plt.show()
```

### More examples on Pie Chart

```python
plt.style.use('ggplot')

slices = [60,40, 30, 20]
labels = ['Sixty','Forty','Thirty','Forty']
colors = ['#235899','#6e1727','#bfd930','#26a324']

plt.pie(slices, labels=labels, colors = colors, wedgeprops={'edgecolor':'black'})

#Wedgeplots - The demarcation between the sectors of the pie chart, preferable
#to set it to black

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

### Real World Problem?
